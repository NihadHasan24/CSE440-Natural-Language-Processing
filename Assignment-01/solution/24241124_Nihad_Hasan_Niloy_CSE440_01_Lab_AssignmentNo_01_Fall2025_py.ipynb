```python
from nltk.corpus import state_union
state_union.fileids()
```

```python
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud

#text file 01
text = state_union.raw('1961-Kennedy.txt')
words = word_tokenize(text)
freq_dist = FreqDist(words)

word_cloud = WordCloud(width = 800, height = 500, background_color= 'white')
cloud = word_cloud.generate_from_frequencies(freq_dist)
plt.figure(figsize = (8,4))
plt.imshow(cloud)
plt.axis('off')
```

**TASK 03**

```python
from nltk.corpus import movie_reviews
movie_reviews.fileids('pos')[0]
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

text = movie_reviews.raw('pos/cv000_29590.txt')
words = word_tokenize(text)

stop_words = stopwords.words('english')
filtered_words = []
for word in words:
  if word.lower() not in stop_words:
    filtered_words.append(word)

lemmatizer = WordNetLemmatizer()
lemmatized_word = []
for word in filtered_words:
  lemmatized_word.append(lemmatizer.lemmatize(word, pos = 'v'))

freq_dist = FreqDist(lemmatized_word)
word = []
count = []
for word_freq in freq_dist.most_common(20):
  word.append(word_freq[0])
  count.append(word_freq[1])

plt.figure(figsize=(18, 6))
plt.bar(word,count)
```

**TASK 04**

```python
from nltk.tokenize import RegexpTokenizer
from nltk.util import ngrams
import numpy as np

text = inaugural.raw('1797-Adams.txt')

tokenizer = RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(text.lower())

stop_words = stopwords.words('english')
filtered_words = []
for word in words:
  if word not in stop_words:
    filtered_words.append(word)

freq_dist = FreqDist(filtered_words)
most_common = freq_dist.most_common(15)
print(most_common)

bigrams = list(ngrams(filtered_words , n = 2))

t_words = []

for w,f in most_common:
  t_words.append(w)

n_bigrams = [biagram for biagram in bigrams if biagram[0] in t_words and biagram[1] in t_words]

word_to_index = {}

for i,word in enumerate(t_words):
  word_to_index[word] = i

matrix = np.zeros((len(t_words),len(t_words)),dtype=int)

for biagram in n_bigrams:
  i = biagram[0]
  j = biagram[1]
  if i in word_to_index and j in word_to_index:
    row = word_to_index[i]
    col = word_to_index[j]
    matrix[row][col]+=1
print(matrix)

```

```python
from nltk.corpus import udhr
udhr.fileids()
```

```python
text = udhr.raw('German_Deutsch-Latin1')
words = word_tokenize(text)

vowels = "AEIOUaeiou"
vowels_seq =[]
for word in words:
  vowel =''
  for i in word:
    if i in vowels:
      vowel += i
  if vowel:
    vowels_seq.append(vowel)

trigrams = list(ngrams(vowels_seq , n = 3))

freq_dist = FreqDist(trigrams)
sort_freq = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)
print(sort_freq)

```
